```python
# Import dependencies.
import matplotlib.pyplot as plt
import statistics
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis, label = "Pyber Boston")
plt.xlabel ('Month')
plt.ylabel ('Fare in USD $')
plt.title ("Fare by Month")
plt.ylim(0, max(y_axis)+5)
plt.legend()
plt.grid()
plt.show()
```

```python
# Create the plot
plt.barh( x_axis,y_axis , color = "green", label = "Pyber Boston")
plt.xlabel ('Month')
plt.ylabel ('Fare in USD $')
plt.title ("Fare by Month")
plt.legend()
plt.grid()
plt.gca().invert_yaxis()
plt.show()
```

```python
plt.scatter(x_axis, y_axis,  marker="x", color = "green", s= y_axis, label = "Pyber Boston")
plt.xlabel ('Month')
plt.ylabel ('Fare in USD $')
plt.title ("Fare by Month")
plt.legend()
plt.grid()
plt.show()
```

```python
make_y_axis_larger = []
for xxxx in y_axis:
  make_y_axis_larger.append(xxxx * 3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
stdev = statistics.stdev(y_axis)
stdev
```

```python
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```
